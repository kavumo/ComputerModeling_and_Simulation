#### Example 2: Exponential Growth	
Suppose that every year, only 10% of fish in a lake have surviving offspring.
If there are currenlty 1000 fish in the lake, plot the graph that will show the projection of the population of the fish in the next 30 years. 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(n):
    return 1.10**n*(1000)
```

```python
x = []
y = []
for n in range(0,31):
    y.append(f(n))
    x.append(n)
```

```python
plt.plot(x,y)
plt.title("Population Vs Years")
plt.ylabel('Population')
plt.ylabel('Years')
```
